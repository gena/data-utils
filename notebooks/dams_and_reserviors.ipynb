```python
# left join dams and reservoirs so that we have a reservoir for every dam
import pandas as pd

dams = fiona.open(path_dams, 'r')
reservoirs = fiona.open(path_reservoirs, 'r')

dams_pd = pd.DataFrame({'id' : [f['properties']['GRAND_ID'] for f in dams], 'dam' : dams})
reservoirs_pd = pd.DataFrame({'id' : [f['properties']['GRAND_ID'] for f in reservoirs], 'reservoir' : reservoirs})

features = pd.merge(dams_pd, reservoirs_pd, how='left', on='id', suffixes=['_left', '_right'])

dams.close()
reservoirs.close()
```

```python
missing = features[[f['properties']['YEAR'] == -99  for f in features['dam']]]
after_1984 = features[[f['properties']['YEAR'] >= 1984 for f in features['dam']]]

print len(missing)
print len(after_1984)
```

```python
# download sattelite photos for the first missing dam

from shapely.geometry import shape

def download_all(ds, path_prefix, max_tile_count=2000):
    # define extent
    for i in range(len(ds)):
        res = ds['reservoir'].iloc[i]
        dam = ds['dam'].iloc[i]

        # print dam['properties']

        if str(res) == 'nan':
            print 'Skipping empty reservoir, ' + str(i)
            continue 
        
        s = shape(res['geometry'])
        s = s.buffer(0.04)
            
        b = s.bounds
        pt1 = ll.transform_coord_geo_to_web((b[0], b[1]))
        pt2 = ll.transform_coord_geo_to_web((b[2], b[3]))

        extent = [pt1[0], pt2[0], pt1[1], pt2[1]]

        year = dam['properties']['YEAR']
        if year > 0:
            start_date='{:04d}-01-01'.format(year - 1)
            end_date='{:04d}-01-01'.format(year + 1)
            tiles = ll.search_tiles(extent, max_cloud = 80, start_date=start_date, end_date=end_date)
        else:
            tiles = ll.search_tiles(extent, max_cloud = 80)

        print(i, len(tiles), start_date)
    
        prefix = path_prefix + '{:06d}_'.format(dam['properties']['GRAND_ID'])

        i = 0
        for t in tiles:
            ll.download_tile_image(t, extent, path_prefix=prefix)
            
            i += 1
            if i > max_tile_count:
                continue
```
